```python
# New project using Google Home Solar panel information from Kaggle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Files/solarstate.csv')
df
```

```python
df.plot(x='lat_max', y='lat_min', kind='scatter')
```

```python
df[['state_name', 'existing_installs_count', 'yearly_sunlight_kwh_total']]
summary = df[['state_name', 'existing_installs_count', 'yearly_sunlight_kwh_total']]
```

```python
summary.sort_values('yearly_sunlight_kwh_total', ascending=False)
```

```python
# pd.merge()
size = pd.read_csv("Files/StateLandSize.csv", usecols=[1,2])
pop = pd.read_csv("Files/StatePop.csv", usecols=[1,2])
size
sizepop = pd.merge(left=size, right=pop)
alldata = pd.merge(left=sizepop, right=summary, right_on='state_name', left_on='State')
alldata.State
```

```python
table = pd.merge(left=alldata, right=df, on='state_name')
table
```

```python
table = table.drop('yearly_sunlight_kwh_total_x', 1)
tabel = table.drop(['state_name', 'region_name', 'lat_max', 'lat_min'], 1)
table = tabel.drop(['install_size_kw_buckets_json', 'carbon_offset_metric_tons', 'existing_installs_count_y' ], 1)
table.rename(columns={'existing_installs_count_x': 'installs', 'yearly_sunlight_kwh_kw_threshold_avg': 'threshold_avg'}, inplace=True)
```

```python
# table = table.drop(['number_of_panels_f', 'number_of_panels_e', 'number_of_panels_w', 'number_of_panels_n', 'number_of_panels_s'], 1)
table
```
