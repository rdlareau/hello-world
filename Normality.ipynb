```python
# Making two variables.
rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

# Sorting the values in ascending order.
rand1.sort()
rand2.sort()

# Making a standard normally distributed variable with 1000 observations,
# a mean of 0, and standard deviation of 1 that we will use as our “comparison.”
norm = np.random.normal(0, 1, 1000)

# Sorting the values in ascending order.
norm.sort()
```

```python
# Plotting the variable rand1 against norm in qqplots.
plt.plot(norm, rand2, "*") 
plt.show() 
```

```python
#Plot a histogram for rand1.
plt.hist(rand1, bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(rand1.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(rand1.mean() + rand1.std()) # , color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(rand1.mean()-rand1.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# Generate a bernoulli distribution with p =0.5.
bernoulli= np.random.binomial(1, .5, 100)

#Plot a histogram.
plt.hist(bernoulli)

# Print the histogram
plt.show()
```

```python
# Generate a binomial distribution with n=20 and p=0.5.
binomial = np.random.binomial(20, 0.5, 1000)

# Plot a histogram.
plt.hist(binomial)

# Print the histogram.
plt.show()
```

```python
# Generate a gamma distribution with shape =5 and scale = 1
gamma = np.random.gamma(20, 4, 1000)

# Plot a histogram.
plt.hist(gamma)

# Print the histogram.
plt.show()
```

```python
expo = np.random.exponential(10, 100)
plt.hist(expo)
plt.show
plt.axvline(expo.mean())
plt.axvline(expo.mean() + expo.std(), linestyle='dashed')
plt.axvline(expo.mean() - expo.std(), linestyle='dashed')
```

```python
logno = np.random.lognormal(0,1,100)
plt.hist(logno)
plt.show
plt.axvline(logno.mean())
plt.axvline(logno.mean() + logno.std(), linestyle='dashed')
plt.axvline(logno.mean() - logno.std(), linestyle='dashed')
```

```python
ray = np.random.rayleigh(20, 100)
plt.hist(ray)
plt.show
plt.axvline(ray.mean())
plt.axvline(ray.mean() + ray.std(), linestyle='dashed')
plt.axvline(ray.mean() - ray.std(), linestyle='dashed')
```

```python
chi = np.random.chisquare(20, 1000)
plt.hist(chi)
plt.show()
plt.axvline(chi.mean())
plt.axvline(chi.mean() + chi.std(), linestyle='dashed')
plt.axvline(chi.mean() - chi.std(), linestyle='dashed')
```

```python
eff = np.random.f(100, 100, 100)
plt.hist(eff)
plt.show
plt.axvline(eff.mean())
plt.axvline(eff.mean() + eff.std(), linestyle='dashed')
plt.axvline(eff.mean() - eff.std(), linestyle='dashed')
```

```python
five = np.random.normal(5, 0.5, 100)
ten = np.random.normal(10, 1, 100)
```

```python
plt.hist(five)
plt.hist(ten)
plt.show
```

```python
tenfive = five + ten
plt.hist(tenfive)
plt.show
plt.axvline(tenfive.mean())
plt.axvline(tenfive.mean() + tenfive.std(), linestyle='dashed')
plt.axvline(tenfive.mean() - tenfive.std(), linestyle='dashed')
```
