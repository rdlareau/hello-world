```python
# For the data plot project using a public data set from Kaggle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read raw csv into pandas (creates 0...x index and all columns)

df = pd.read_csv('Files/India.csv')
df
```

```python
# Plot of the raw unstorted values

plt.plot(df.value)
```

```python
# Select import entries from one specific country

df[df["country_name"] == "Tuvalu"]
```

```python
# Select entries which an entry value > 1000, sum the totals by country, and sort
# Note that columns which cannot be summed are excluded from the output
# Generating sums of the "country code" is meanlingless

summary = df[df.value >= 1000].groupby('country_name').sum().sort_values("value", ascending=False)
summary
```

```python
plt.plot(summary['value'])
```

```python
s = pd.DataFrame(summary)
s.index.name = 'country_name'
s.index.values
s['value']
```

```python
plt.plot(s.index.values, s['value'])
```
